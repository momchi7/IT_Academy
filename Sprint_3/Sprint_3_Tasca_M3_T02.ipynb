Fes un informe complet del dataset:

```python
delayed_flights_modified["Date"]=pd.to_datetime(dict(
    year=delayed_flights_modified["Year"],
    month=delayed_flights_modified["Month"],
    day=delayed_flights_modified["DayofMonth"]))
```

```python
# set column "Date" as index. 
delayed_flights_modified.set_index("Date", inplace=True)
```

```python
delayed_flights_modified
```

Resumeix estadísticament el dataset i les columnes d’interès. Fes una anàlisi estadístic del que consideris rellevant.

```python
delayed_flights_modified[["ActualElapsedTime","DepDelay","AirTime","ArrDelay",
                          "DepDelay","Distance"]].describe()
```

We can observe in the above table that there's a big difference between the **mean** and **median** values for each variable. This could indicate the presence of extreme values, which tend to pull the mean i.e. the maximum delay for departures is 1952 minutes (32.5 hours). Therefore, the median would be more representative. 



Troba quantes dades faltants hi ha per columna.

```python
# ArrTime, ActualElapsedTime, CRSElapsedTime, AirTime, ArrDelay, TaxiIn, TaxiOut

delayed_flights_modified.isna().sum()
```

Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...).

```python
#velocitat mitjana del vol

air_time_in_hours = delayed_flights_modified["AirTime"]/60

delayed_flights_modified["Velocity"]=delayed_flights_modified["Distance"]/air_time_in_hours

delayed_flights_modified.groupby(by="FlightNum")["Velocity"].mean().round()
```

Mean velocity for FlightNum 5 is 421km/h.

```python
delayed_flights_modified.groupby(by="FlightNum")["Velocity"].mean().round().nlargest()
```

**Top 5** flights with the fastest mean velocity, i.e. flight number **5158** has a mean velocity of **711 km/h**.

```python
# si ha arribat tard o no
# ArrDelay, arrival delay in minutes

delayed_flights_modified["LateArrival"]=delayed_flights_modified["ArrDelay"]>0

delayed_flights_modified[["FlightNum","LateArrival"]]
```

Fes una taula de les aerolínies amb més endarreriments acumulats.

```python
delayed_flights_modified["LateDeparture"] = delayed_flights_modified["DepDelay"]>0

airline_flight_delays= delayed_flights_modified.groupby(by="UniqueCarrier")["LateArrival","LateDeparture"].sum()

airline_flight_delays["AccumulatedDelays"] = airline_flight_delays["LateArrival"] + airline_flight_delays["LateDeparture"]

airline_flight_delays.sort_values(by=["AccumulatedDelays"], ascending=False)
```

The airline with the highest number of accumulated delays is **WN**, **71971**.

Quins són els vols més llargs? I els més endarrerits? Busca les rutes més llargues i les que acumulen més retards.

```python
# top 10 vuelos más largos

delayed_flights_modified[["FlightNum", "AirTime"]].sort_values(by="AirTime", ascending=False).nlargest(10,columns="AirTime")
```

El vuelo **#15** del 19 de octubre fue el más largo, **654 minutos**.
